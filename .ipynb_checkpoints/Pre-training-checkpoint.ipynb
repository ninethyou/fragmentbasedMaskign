```python
import csv
```

```python
data_path1=  'data/pretrain_example.txt'
data_path2=  'data/pubchem-10m-clean.txt'
```

```python
smiles_data_small = []
with open(data_path1) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for i, row in enumerate(csv_reader):
        smiles = row[-1]
        smiles_data_small.append(smiles)
```

```python
smiles_data_large = []
with open(data_path2) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for i, row in enumerate(csv_reader):
        smiles = row[-1]
        smiles_data_large.append(smiles)
```
