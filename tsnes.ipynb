```python
import matplotlib.pyplot as plt
import numpy as np

# 예시 데이터 생성 (실제 데이터로 교체하세요)
np.random.seed(42)
test_latent = np.random.randn(100, 2)
test_y = np.random.randint(0, 2, 100)

# 데이터 그리기
plt.figure(figsize=(10, 10))
plt.scatter(test_latent[test_y == 0, 0], test_latent[test_y == 0, 1], label='클래스 0', alpha=0.7)
plt.scatter(test_latent[test_y == 1, 0], test_latent[test_y == 1, 1], label='클래스 1', alpha=0.7)
plt.title('이진 분류 잠재 공간', fontsize=15)
plt.xlabel('Z1', fontsize=15)
plt.ylabel('Z2', fontsize=15)
plt.legend(fontsize=15)
plt.axis('equal')
plt.show()
```
