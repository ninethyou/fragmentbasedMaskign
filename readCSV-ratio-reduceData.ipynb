```python
import os
import pandas as pd 

results_df = pd.DataFrame()  # 결과를 저장할 빈 데이터 프레임 생성

for dataset in selected_datasets_lower:
    file_name = 'pretrained_gin_{}_scaffold_random_masking_mask_edge{}.csv'.format(dataset,file)
    file_path = os.path.join(base_path ,file_name)
    
    
    if os.path.isfile(file_path):
        print(file_path)
        
        temp_df  = pd.read_csv(file_path)

        grouped = temp_df.groupby('reduceData')['result'].agg(['mean', 'std']).reset_index()
        
        grouped['Dataset'] = dataset
        
        # 결과 데이터 프레임에 추가
        results_df = pd.concat([results_df, grouped], ignore_index=True)
    else:
        print(f"{file_path} 파일이 존재하지 않습니다.")

# 결과 데이터 프레임의 열 이름 변경
results_df.columns = ['mask_rate', 'Mean Result', 'Stddev Result', 'Dataset']

# 최종 결과 출력
print(results_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming `results_df` is the DataFrame with your data
df = pd.DataFrame(results_df)

# Define which datasets are classification and which are regression
classification_datasets = ['sider', 'tox21', 'bbbp', 'bace']
regression_datasets = list(set(df['Dataset']) - set(classification_datasets))

# Plot classification datasets together
plt.figure(figsize=(10, 6))
classification_data = df[df['Dataset'].isin(classification_datasets)]
for name, group in classification_data.groupby('Dataset'):
    plt.plot(group['mask_rate'], group['Mean Result'], marker='o', label=name)
plt.title('Mean Result by Mask Rate for Classification Datasets')
plt.xlabel('Mask Rate')
plt.ylabel('Mean Result')
plt.grid(True)
plt.legend()
plt.show()

# Plot freesolv dataset separately
freesolv_data = df[df['Dataset'] == 'freesolv']
plt.figure(figsize=(10, 6))
plt.plot(freesolv_data['mask_rate'], freesolv_data['Mean Result'], marker='o', label='freesolv', color='green')
plt.title('Mean Result by Mask Rate for freesolv Dataset')
plt.xlabel('Mask Rate')
plt.ylabel('Mean Result')
plt.grid(True)
plt.legend()
plt.show()


# Plot freesolv dataset separately
qm7_data = df[df['Dataset'] == 'qm7']
plt.figure(figsize=(10, 6))
plt.plot(qm7_data['mask_rate'], qm7_data['Mean Result'], marker='o', label='qm7', color='blue')
plt.title('Mean Result by Mask Rate for freesolv Dataset')
plt.xlabel('Mask Rate')
plt.ylabel('Mean Result')
plt.grid(True)
plt.legend()
plt.show()

# Plot other regression datasets (excluding classification datasets and freesolv)
plt.figure(figsize=(10, 6))
other_regression_data = df[(df['Dataset'].isin([ 'esol', 'lipo']))]
for name, group in other_regression_data.groupby('Dataset'):
    plt.plot(group['mask_rate'], group['Mean Result'], marker='o', label=name)
plt.title('Mean Result by Mask Rate for Other Regression Datasets (Excluding freesolv)')
plt.xlabel('Mask Rate')
plt.ylabel('Mean Result')
plt.grid(True)
plt.legend()
plt.show()
```
